```python
#This section creates two libraries (sets) of harmful chemicals: 
#one for carcinogens and one for reproductive toxicants.

carcinogens = {
    'Formaldehyde', 'Titanium dioxide', 'Carbon black', 'Benzene', 
    'Benzyl chloride', 'Vinyl acetate', '1,4-Dioxane', 'Acrylamide', 
    'Styrene', 'Benzo[a]pyrene', 'Benz[a]anthracene', 'Coal tar', 
    'Coal tar extract', 'Safrole', 'Methyleugenol', 'Ethyl acrylate',
    'Talc', 'N-Nitrosodimethylamine', 'N-Nitrosodiethanolamine'
}

reproductive_toxicants = {
    'Lead acetate', 'Lead', 'Mercury', 'Mercury compounds', 
    'Di-n-butyl phthalate', 'Retinol', 'Retinyl esters', 
    'Vitamin A palmitate', 'Retinyl acetate', 'Bisphenol A', 
    'Ethylene oxide', 'Propylene oxide', 'Cocamide DEA', 
    'Diethanolamine', 'Selenium sulfide', 'Spironolactone'
}
```

```python
#This function checks if a chemical is in either harmful chemical library
#and returns True if it is, otherwise False.


def toxin (row):
    if row['ChemicalName'] in carcinogens:
        return True
    elif row['ChemicalName'] in reproductive_toxicants:
        return True
    else:
        return False
```

```python
#This code applies the toxin function to each row in the DataFrame and creates a
#new column 'toxin' that indicates whether the product contains a harmful chemical.


cscpopendata_df['toxin'] = cscpopendata_df.apply(toxin, axis = 1)
cscpopendata_df.head()
```

```python
#Counts how many products are flagged as toxins or not.

cscpopendata_df['toxin'].value_counts()
```

```python
#Classifies chemicals into groups and adds the result in a new column.


def classify_chemical(name):
    if any(c.lower() in name for c in carcinogens):
        return 'carcinogens'
    elif any(c.lower() in name for c in reproductive_toxicants):
        return 'reproductive_toxicants'
    else:
        return 'Other/Unclassified'

cscpopendata_df['ChemicalGroup'] = cscpopendata_df['ChemicalName'].apply(classify_chemical)
```

```python
#create a new data frame with a category for carcinogens and reproductive_toxicants

high_risk = cscpopendata_df[cscpopendata_df['ChemicalGroup'].isin(['carcinogens', 'reproductive_toxicants'])]
```

```python
#Defines a list of specific product names from the "Too Faced Cosmetics" brand that has high-risk chemicals.


target_products = [
    'Too Faced Unicorn Survival Kit Mystical & Lip Highlighter Set',
    'Too Faced Sweet Peach Eye Shadow Palette',
    'Too Faced White Peach Multi-Dimensional Eye Shadow Palette',
    "Too Faced Papa Don't Peach",
    'Too Faced Sweetie Pie Radiant Matte Bronzer'
]
```

```python
#Filters the dataset to include only products that have been discontinued.

discontinued_df= cscpopendata_df[cscpopendata_df['DiscontinuedDate'].notnull()]
```

```python
#Selects discontinued products from the target list and retrieves their unique names.


discontinued_products = (
    discontinued_df[discontinued_df['ProductName'].isin(target_products)]
    ['ProductName']
    .unique()
)
```

```python
#Creates a list that records whether each target product is discontinued or still active.


status_results = []
for product in target_products:
    status = "Discontinued" if product in discontinued_products else "Active"
    status_results.append({"ProductName": product, "Status": status})
```

```python
#Converts the product status list into a DataFrame.


status_df = pd.DataFrame(status_results)
status_df
```
